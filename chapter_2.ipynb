```python
import numpy as np

def is_basis(vectors):
    """
    Kiểm tra xem tập vectors có phải là một cơ sở của không gian hay không.
    """
    # Chuyển danh sách vector thành ma trận
    matrix = np.array(vectors).T  # Chuyển vector thành cột
    
    # Lấy số vector và số chiều của không gian
    num_vectors = matrix.shape[1]
    dimension = matrix.shape[0]
    
    # Kiểm tra: Số vector phải bằng số chiều của không gian
    if num_vectors != dimension:
        return False, "Số vector không bằng số chiều của không gian => Không phải cơ sở."
    
    # Kiểm tra tính tuyến tính của ma trận bằng số chiều không gian
    rank = np.linalg.matrix_rank(matrix)
    if rank < dimension:# nếu hệ vector có ít hơn n vector, thì nó không thể sinh ra toàn bộ không gian R^n
        return False, "Tập vector phụ thuộc tuyến tính => Không phải cơ sở."
    
    return True, "Tập vector là một cơ sở của không gian."

# Nhập số chiều của không gian
try:
    n = int(input("Nhập số chiều của không gian vector (n): "))
    if n <= 0:
        raise ValueError
except ValueError:
    print("Lỗi! n phải là số nguyên dương!")
    exit()

# Nhập các vector từ bàn phím
vectors = []
print(f"Nhập {n} vector, mỗi vector gồm {n} phần tử (cách nhau bởi dấu cách):")
for i in range(n):
    try:
        vector = list(map(float, input(f"Vector [{i+1}]: ").split()))
        if len(vector) != n:
            raise ValueError
        vectors.append(vector)
    except ValueError:
        print(f"Lỗi! Mỗi vector phải đúng {n} phần tử!")
        exit()

print(vectors)

is_basis_result, message = is_basis(vectors)
print(message)
```

```python
import numpy as np

class NormedSpace:
    def __init__(self, vector):
        self.vector = np.array(vector)

    def l1_norm(self):
        """Chuẩn L1 (Manhattan)"""
        return np.sum(np.abs(self.vector))

    def l2_norm(self):
        """Chuẩn L2 (Euclidean)"""
        return np.round(np.sqrt(np.sum(self.vector ** 2)),2)

    def linf_norm(self):
        """Chuẩn L∞ (Infinity)"""
        return np.max(np.abs(self.vector))

    def check_triangle_inequality(self, other_vector):
        """
        Kiểm tra bất đẳng thức tam giác: ||x + y|| ≤ ||x|| + ||y||
        """
        other_vector = np.array(other_vector)
        norm_x = np.linalg.norm(self.vector, ord=2)
        norm_y = np.linalg.norm(other_vector, ord=2)
        norm_x_plus_y = np.linalg.norm(self.vector + other_vector, ord=2)
        return norm_x_plus_y <= norm_x + norm_y

    # Nhập vector từ bàn phím
    def get_vector_from_input():
        vector_str = input("Nhập vector (các phần tử cách nhau bằng dấu cách): ")
        vector = list(map(float, vector_str.split()))
        return vector


# Nhập hai vector từ bàn phím
print("Nhập vector x:")
x = NormedSpace.get_vector_from_input()
print(f"Vector x: {x}")

print("Nhập vector y:")
y = NormedSpace.get_vector_from_input()
print(f"Vector y: {y}")

norm_x = NormedSpace(x)
norm_y = NormedSpace(y)

# Tính các chuẩn
print(f"Chuẩn L1 của x: {norm_x.l1_norm()}")
print(f"Chuẩn L2 của x: {norm_x.l2_norm()}")
print(f"Chuẩn L∞ của x: {norm_x.linf_norm()}")

# Kiểm tra bất đẳng thức tam giác
print("Kiểm tra bất đẳng thức tam giác:", end=" ")
if norm_x.check_triangle_inequality(norm_y.vector):
    print("Đúng")
else:
    print("Sai")
```

```python
def rank_of_vectors(dim,num_vectors):
    """
    Tính hạng của một hệ vector trong R^n
    :param dim: Chiều không gian của vector (số phần tử trong mỗi vector)
    :param num_vectors: Số lượng vector trong hệ
    :return: Hạng của hệ vector
    """
    vectors = []

    print(f"Nhập {num_vectors} vector trong không gian R^{dim}")
    for i in range(num_vectors):
        hung = list(map(float,input(f"\nNhập v{i+1}:").split()))
        if len(hung) != dim:
            print(f"Lỗi: khi nhập v{i+1} !")
            return
        vectors.append(hung)
    matrix = np.array(vectors)
    rank = np.linalg.matrix_rank(matrix)
    print(f"\nMa trận ta có:\n {matrix}")
    print(f"\nHạng của hệ vector là : {rank}")
    print(f"\nMa trận khi chuyển mỗi vector thành cột:\n {np.column_stack(matrix)}")
    print(f"\nHạng ma trận chuyển: {np.linalg.matrix_rank(np.column_stack(matrix))}")
dim = int(input("Nhập chiều không gian của vector (n): "))
num_vectors = int(input("Nhập số vector trong hệ"))
rank_of_vectors(dim,num_vectors)  
```

```python
import numpy as np

def gram_schmidt(V):
    """
    Thực hiện trực chuẩn hóa Gram-Schmidt trên tập hợp vector V.

    Args:
        V: Ma trận numpy (m x n), mỗi cột là một vector đầu vào.

    Returns:
        U: Ma trận chứa các vector trực chuẩn.
    """
    V = np.zeros(V.shape, dtype=np.float64)
    m,n = V.shape  
    U = []
    for i in range(n):
        u_i = V[:, i]  # Lấy vector hiện tại

        for u_j in U:
            # Loại bỏ thành phần chiếu lên các vector trước đó
            proj = (np.dot(u_j, u_i)/np.dot(u_j,u_j)) * u_j
            u_i = u_i - proj

        
        norm_u_i = np.linalg.norm(u_i) # Chuẩn của vector
        if norm_u_i > 1e-10:  # Kiểm tra nếu vector không bị triệt tiêu hoàn toàn
            U.append(u_i/norm_u_i)  # Chuẩn hóa vector

    return np.array(U).T  # Chuyển dạng ma trận để mỗi cột là một vector trực chuẩn

# Nhập vector từ người dùng
num_vectors = int(input("Nhập số chiều của vector: "))
vector_dim = int(input("Nhập số lượng vector: "))

vectors = []
print("Nhập từng vector (các phần tử cách nhau bằng dấu cách):")
for i in range(num_vectors):
    vector = list(map(float, input(f"Vector {i+1}: ").split()))
    if len(vector) != vector_dim:
        print("Lỗi: Số phần tử không đúng, vui lòng nhập lại!")
        exit()
    vectors.append(vector)
# chuyển đổi sang numpy array và thực hiện Gram-Schimidt
V = np.array(vectors).T  # Chuyển ma trận để mỗi cột là một vector

# Áp dụng Gram-Schmidt
U = gram_schmidt(V)

print("\nCác vector trực chuẩn hóa:")
print(U.T)  # Chuyển lại dạng vector ban đầu
```

```python
import numpy as np

def is_linear_mapping(A):
    """ Kiểm tra ảnh của ma trận có phải là ánh xạ tuyến tính không. """
    return True  # Mọi ánh xạ biểu diễn bởi ma trận đều là tuyến tính

def find_image(A):
    """ Tìm ảnh của ánh xạ tuyến tính (Im(f)) (tức là không gian cột của ma trận A). """
    U, S, Vt = np.linalg.svd(A)  # Phân rã SVD để tìm các vector độc lập
    rank = np.linalg.matrix_rank(A)  # Hạng của A
    image_basis = U[:rank]  # Cơ sở của không gian cột (Imf)
    return image_basis

def find_kernel(A):
    """ Tìm hạt nhân của ánh xạ tuyến tính (Ker(f)), tức là tập nghiệm của Ax = 0. """
    m,n = A.shape
    U, S, Vt = np.linalg.svd(A)  # Phân rã SVD để tìm không gian null
    null_space = Vt[len(S):].T  # Vector tương ứng với giá trị suy biến
    return null_space

# Nhập kích thước của ma trận ánh xạ tuyến tính
m = int(input("Nhập số hàng của ma trận ánh xạ: "))  # Số chiều đích
n = int(input("Nhập số cột của ma trận ánh xạ: "))  # Số chiều miền xác định

# Nhập ma trận ánh xạ tuyến tính
print(f"Nhập ma trận phản ánh tuyến tính {m}x{n} (mỗi hàng nhập trên 1 dòng, các phần tử cách nhau dấu cách):")
A = np.array([list(map(float, input().split())) for _ in range(m)])

# Kiểm tra ánh xạ có phải tuyến tính không
if is_linear_mapping(A):
    print("Ánh xạ f là tuyến tính.")

# Tìm ảnh của f (Im(f))

image = find_image(A)
print("Ảnh của f (Im(f)):")
print(image)

# Tìm hạt nhân của f (Ker(f))
Kernel = find_kernel(A)
print("Hạt nhân của f (Ker(f)):")
print(Kernel if Kernel.size > 0 else "(không có vector nào)")
# tính hạng và số khuyết
# Tính nullity
rank = np.linalg.matrix_rank(A)
nullity = A.shape[1] - rank  # Số khuyết = số biến - hạng
print(f"\nhạng của f (rank): {rank}")
print(f" Số khuyết của f (nullity): {nullity}")

# Nhập vector cần ánh xạ
print(f"Nhập vector đầu vào để ánh xạ (kích thước phải là {n}x1):")
vector = np.array(list(map(float, input().split()))).reshape(n, 1)

# Kiểm tra kích thước hợp lệ
if vector.shape[0] != n:
    print("\nLỗi: Kích thước vector đầu vào không hợp lệ với miền xác định!")
else:
    # Tính ảnh của vector
    result = A @ vector #Tính f(vector) = A * vector
    print("\nKết quả ảnh xạ của vector:")
    print(result)

```
